```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.metrics import mean_squared_error
```

```python
# Load data for all images, and for test targets and predictions
df_targets = pd.read_csv('./targets.csv', index_col=0)
df_true = pd.read_csv('./test_targets.csv', index_col=0)
df_predictions = pd.read_csv('./test_predictions.csv', index_col=0)
```

```python
# Set column names
df_predictions.columns = df_targets.columns
df_true.columns = df_targets.columns
```

```python
# Check total MSE between datasets for verification purposes
mean_squared_error(df_true.values, df_predictions.values)
```

```python
# Create dataframe of MSE for each class for further analysis
df_mse = pd.DataFrame(index = df_true.columns.values, columns=['class_MSE'])
for c in df_true.columns:
    df_mse.at[c, 'class_MSE'] = mean_squared_error(df_true[c].values, df_predictions[c].values)
```

```python
df_mse['class_MSE'] = df_mse['class_MSE'].astype(float)
```

```python
#Plot maximum predicted agreement per class
plt.subplots(figsize=(12,6))
sns.barplot(x=df_predictions.columns, y=df_predictions.max(),
            order=df_predictions.max().sort_values(ascending=False).index,
            palette='mako')

# Formatting
plt.xticks(rotation=80)
plt.grid(axis='y', color='white')
plt.title('Maximum Predicted Agreement Per Class', fontsize=14)
plt.ylabel('Maximum Percent Predicted Agreement', fontsize=12)

plt.show()
```

```python
# Plot predicted MSE by class
df_mse = df_mse.sort_values(by='class_MSE', ascending=False)
fig, ax = plt.subplots(figsize=(12,8))
sns.barplot(x=df_mse.index, y=df_mse['class_MSE'], palette='bone')

# Formatting
plt.xticks(rotation=90)
plt.ylabel('Class Mean Squared Error', fontsize=12)
plt.title('Predictions Mean Squared Error by Class', fontsize=16)
plt.show()
```

```python
#Create a dataframe to compare predicted versus actual class frequencies
df_freq = pd.DataFrame(index=df_true.columns, columns=['avg_freq_true', 'avg_freq_predicted'])
df_freq['avg_freq_true'] = df_true.mean().values
df_freq['avg_freq_predicted'] = df_predictions.mean().values
```

```python
#Evaluate how predicted class frequencies relate to MSE
stats.pearsonr(df_freq['avg_freq_predicted'], df_mse['class_MSE'])
```

```python
sns.jointplot(x='avg_freq_predicted', y='avg_freq_true', data=df_freq, kind='reg', height=8)
plt.xlabel('Average Predicted Class Frequency', fontsize=12)
plt.ylabel('Average Actual Class Frequency', fontsize=12)
plt.show()
```

```python
stats.pearsonr(df_freq['avg_freq_predicted'], df_freq['avg_freq_true'])
```

```python
# Select class images above percent agreement threshold
threshold_pct = 0.8
e_filt = df_predictions['Class1.1'] > threshold_pct
s_filt = (df_predictions['Class4.1'] > threshold_pct) & (df_predictions['Class3.2'] > threshold_pct)
sb_filt = (df_predictions['Class4.1'] > threshold_pct) & (df_predictions['Class3.1'] > threshold_pct)

# Get filenames corresponding with selected images
e_galaxy_fn = './images/' + str(df_predictions[e_filt].iloc[0].name) + '.jpg'
s_galaxy_fn = './images/' + str(df_predictions[s_filt].iloc[0].name) + '.jpg'
sb_galaxy_fn = './images/' + str(df_predictions[sb_filt].iloc[0].name) + '.jpg'

# Print galaxy numbers of each (files to be copied into image directory manually)
print('Elliptical class: {}, Spiral class {}, Spiral-bar class {}'.format(df_predictions[e_filt].iloc[0].name,
                                                                          df_predictions[s_filt].iloc[0].name,
                                                                          df_predictions[sb_filt].iloc[0].name))
                                                                          
```

```python
# Read image data from each file 
e_img = plt.imread(e_galaxy_fn)
s_img = plt.imread(s_galaxy_fn)
sb_img = plt.imread(sb_galaxy_fn)

# Set up image grid
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(16, 6))

# Load image data (read above) for display
ax1.imshow(e_img)
ax2.imshow(s_img)
ax3.imshow(sb_img)

# Turn off axes
ax1.axis('off')
ax2.axis('off')
ax3.axis('off')

# Set plot titles
ax1.set_title('Elliptical', fontsize=14)
ax2.set_title('Spiral', fontsize=14)
ax3.set_title('Spiral Bar', fontsize=14)

# Set overall title
fig.suptitle ('General Galaxy Classes with Over 80% Agreement', fontsize=18)

plt.tight_layout()
plt.show()
```
