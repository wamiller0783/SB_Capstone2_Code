### Detailed Analysis

Before proceding, it is worth noting that the data classes listed above are based on crowd-sourced data. When the data was crowd-sourced, those labelling the data were given the tasks laid out below, with each task potentially leading to other tasks, as indicated in the "Next" column. Each class above corresponds to task and an answer below. In other words an answer of "No" on task 3 would be assigned Class 3.2.

<img src="./GalaxyClasses.png">

The above chart can, to at least some degree, be distilled into Hubble's galaxy classification system. As can be seen below, this divides galxies into ellipticals, spirals, and spiral bar classes.
<ul>
    <li>Ellipticals would be indicated by class 1.1 in the above, though this may not be true if it is accompanied by a high classification percentage on class 6.1.</li>
    <li>Spirals would be indicated by classes 4.1 and 3.2 together (again, with 6.1 possibly effecting it).</li>
    <li>Spiral bar galaxies would be indicated by classes 4.1 and 3.1 together (again, with 6.1).</li>
</ul>
Subclasses of each of these would be indicated in classes 7 (degrees of elliptical), or 10 and 11 (types of spiral and spiral bar galaxies).

<img src="./HubbleTuningFork.jpg" width=400>

Keeping the above in mind, let's take a look at which categories tend to be assigned more frequently than others with a high degree of agreement among data labellers.

```python
df_plt = df.melt(var_name='classes', value_name='percent')
fig, ax = plt.subplots(figsize=(16,8))
# Calculate percentage of each class with high agreement (over 66%)
agreement_level_high = (df_plt[df_plt['percent'] >= 0.66]['classes'].value_counts()/
                   df_plt['classes'].value_counts().dropna())
# Calculate percentage of each class with majority agreement (over 50%)
agreement_level_maj = (df_plt[df_plt['percent'] > 0.5]['classes'].value_counts()/
                   df_plt['classes'].value_counts().dropna())

# Plot overlapping barplots for the two groups above
sns.barplot(x=agreement_level_maj.index, 
            y=agreement_level_maj.values,
            order=agreement_level_maj.sort_values(ascending=False).index,
            palette='Oranges', alpha = 1,
            ax=ax, label='Majority Agreement (>50%)')
sns.barplot(x=agreement_level_high.index, 
            y=agreement_level_high.values,
            order=agreement_level_maj.sort_values(ascending=False).index,
            palette='Blues_r', alpha = 0.75,
            ax=ax, label= 'High Agreement (>66%)')

# Formatting
plt.xticks(rotation=80)
plt.title('Class Distribution by Majority Agreement', fontsize=14)
plt.ylabel('Percent Agreement', fontsize=12)
plt.legend()
plt.show()
```

We can get additional information about the above by looking at a box plot to see the median values and quartile ranges of agreement for each class distribution.

```python
# Plot class distr
plt.subplots(figsize=(16,8))
sns.boxplot(x='classes', y='percent', data=df_plt, 
            order=df.median().sort_values(ascending=False).index, fliersize=0)
plt.title('Class Distribution by Percent Agreement', fontsize=14)
plt.xlabel('Classes', fontsize=12)
plt.ylabel('Percent Agreement', fontsize=12)
plt.xticks(rotation=80)
plt.show()
```

Important information that can be gleaned from this, from left to right:
<ol>
    <li>There is nothing "odd" about the majority of labelled galaxy images. (Class 6.2 compared to 6.1)</li>
    <li>More than half of galaxy images include some kind of feature or disk, but a significant range is observed in the every quartile (Class 1.1)</li>
    <li>Around 40% of galaxies could *not* be disks viewed edge-on, though significant range is oberved in every quartile. (Class 2.2 compared to 2.1)</li>
    <li>Where galaxies are not disks viewed edge-on, most tend not have have a bar (Class 3.2 compared to 3.1)
</ol>

```python
plt.subplots(figsize=(12,6))
sns.barplot(x=df.columns, y=df.max(),
            order=df.max().sort_values(ascending=False).index,
            palette='mako')
plt.xticks(rotation=80)
plt.grid(axis='y', color='white')
plt.title('Maximum Agreement Per Class', fontsize=14)
plt.ylabel('Maximum Percent Agreement', fontsize=12)

plt.show()
```

Because some tasks in the data labelling process lead to others, we would expect to see a strong measure of correlation between specific classes.
This should be observable in a correlation matrix.

```python
plt.subplots(figsize=(12,12))
sns.heatmap(df.corr(), cmap='Greys_r')
plt.show()
```

Looking closely at the above, we find about what we would expect. The starting and ending tasks (classes 1 and 6) with a “yes” or “no” distinction are clearly negatively correlated. The other “yes” and “no” classes vary since they are dependent on other tasks. Other classes can be evaluated one-by-one to see that they make since. For instance, since the first answer on task one would send a crowd-sourced data labeller on to task 7, Class 1.1 correlates highly with classes 7.1 and 7.2. As could be seen in the description of the dataframe, class 7.3 was assigned rather uncommonly (with a mean value of 0.05), which explains the lesser degree of correlation there. Stepping through each class similarly, we can observe the patterns defined by the data-labelling task list. 

To finish up the exploring this data, it would be interesting to see if galaxies that are classified with a high degree of agreement visually correspond with their Hubble classification.

```python
# Select class images above percent agreement threshold
threshold_pct = 0.9
e_filt = df['Class1.1'] > threshold_pct
s_filt = (df['Class4.1'] > threshold_pct) & (df['Class3.2'] > threshold_pct)
sb_filt = (df['Class4.1'] > threshold_pct) & (df['Class3.1'] > threshold_pct)

# Get filenames corresponding with selected images
e_galaxy_fn = './images/' + str(df[e_filt].iloc[0].name) + '.jpg'
s_galaxy_fn = './images/' + str(df[s_filt].iloc[0].name) + '.jpg'
sb_galaxy_fn = './images/' + str(df[sb_filt].iloc[0].name) + '.jpg'

# Print galaxy numbers of each (files to be copied into image directory manually)
print('Elliptical class: {}, Spiral class {}, Spiral-bar class {}'.format(df[e_filt].iloc[0].name,
                                                                          df[s_filt].iloc[0].name,
                                                                          df[sb_filt].iloc[0].name))
                                                                          
```

```python
# Read image data from each file 
e_img = plt.imread(e_galaxy_fn)
s_img = plt.imread(s_galaxy_fn)
sb_img = plt.imread(sb_galaxy_fn)

# Set up image grid
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(16, 6))

# Load image data (read above) for display
ax1.imshow(e_img)
ax2.imshow(s_img)
ax3.imshow(sb_img)

# Turn off axes
ax1.axis('off')
ax2.axis('off')
ax3.axis('off')

# Set plot titles
ax1.set_title('Elliptical', fontsize=14)
ax2.set_title('Spiral', fontsize=14)
ax3.set_title('Spiral Bar', fontsize=14)

# Set overall title
fig.suptitle ('General Galaxy Classes with Over 90% Agreement', fontsize=18)

plt.tight_layout()
plt.show()
```

```python
# Check what percentage of each class has a 90% or higher agreement
galaxy_types = {'elliptical': e_filt, 'spiral': s_filt, 'spiral-bar': sb_filt}
for key in galaxy_types:
    print('There are {:0.3f}% galaxies classified as {} with over 90% agreement.'.format(galaxy_types[key].sum()/len(df), key))
```
